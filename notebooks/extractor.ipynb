```python
from pypdf import PdfReader

def print_pdf_text(pdf_path: str) -> None:
    """
    Open the PDF at pdf_path and print the text content of each page.
    """
    # Create a PdfReader object to read the file
    reader = PdfReader(pdf_path)  # uses PdfReader under the hood :contentReference[oaicite:3]{index=3}

    # Iterate over all pages and extract text
    for page_number, page in enumerate(reader.pages, start=1):
        text = page.extract_text()
        print(f"\n--- Page {page_number} ---\n")
        print(text)

if __name__ == "__main__":

    pdf_file = "../documents/data.pdf"
    print_pdf_text(pdf_file)
```
